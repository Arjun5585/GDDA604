```python
import matplotlib.pyplot as plt

# Data
status = ['Cancelled', 'Disputed', 'In Process', 'On Hold', 'Resolved', 'Shipped']
total_sales = [194487.48, 72212.86, 144729.96, 178979.19, 150718.28, 9291501.08]

# Calculate total and percentage contribution
total = sum(total_sales)
percentages = [(sales / total) * 100 for sales in total_sales]

# Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(status, total_sales, color='skyblue')

# Add percentage labels above bars
for bar, percentage in zip(bars, percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{percentage:.2f}%', ha='center', va='bottom', fontsize=10)

# Add titles and labels
plt.title('Total Sales by Order Status', fontsize=14)
plt.xlabel('Order Status', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Display the plot
plt.show()
```

```python
#Null Hypothesis (H₀): There is no significant difference in mean sales between North America and Europe.
#Alternative Hypothesis (H₁): There is a significant difference in mean sales between North America and Europe.
```

```python
from scipy import stats
# Calculate the means of both regions
mean_sales_na = NA_sales_df.mean()
mean_sales_eu = EU_sales.mean()

print(f"Mean sales in North America: {mean_sales_na}")
print(f"Mean sales in Europe: {mean_sales_eu}")


# Assuming 'NA_sales' and 'EU_sales' are pandas Series for sales data of North America and Europe
# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(NA_sales, EU_sales)

# Define significance level
alpha = 0.05

# Hypothesis Test Result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Define price categories based on the specified price ranges
bins = [0, 40, 79, float('inf')]  # Less than 40, 41-79, More than 80
labels = ['low', 'medium', 'high']

# Create a new column for price category
data['PriceCategory'] = pd.cut(data['PRICEEACH'], bins=bins, labels=labels, right=False)

# Group by the price category and calculate the sum of QUANTITYORDERED for each category
quantity_by_price_category = data.groupby('PriceCategory')['QUANTITYORDERED'].sum()
```
